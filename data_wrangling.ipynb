### QUESTION 1


### Load and read in personality Data Frame

```python
df_scores = pd.read_csv("personality_scores.csv", sep=";")
```

```python
#checking for duplicates
dup = df_scores[df_scores.duplicated(['ID'])]
print(dup)
```

### Dropping null values

```python
#Dropping duplicates and NAN values
df_scores = df_scores.dropna(axis= 1, how='all')
df_scores.head(5)
```

### Assert statement to check that the new data frame is the length of the unique entries of the original data frame.

```python
#Assertation
try:
    assert df_scores['ID'].unique().sum() == df_scores['ID'].sum(), "the new data frame is not equal to the length of the unique entries of the original data frame"
    print('True')
except AssertionError as msg:
    print(msg)
```

## QUESTION 2

### Total score of each of the personality test subscales

```python
# Making dictionaries
category = dict({1:'Extraversion', 2:'Agreeableness', 3:'Conscientiousness', 4:'Emotional Stability', 5:'Intellect'})
```

```python
import re

#Creating new columns containing the total score of each of the personality test subscales.
#calculating the total score for each of the subscales
def score(entry, category):
    numbers=0
    _r= re.compile('\d+')
    for col in entry:
        num= _r.findall(col)
        if str(category)==num[0]:
            numbers+= int(num[1])
        else:
            numbers= numbers
    return numbers
    
for i in category.keys():
    df_scores[f'{category[i]}']= df_scores.loc[:,'Section 5 of 6 [I am always prepared.]':'Section 5 of 6 [I worry about things.]'].apply(score, args=(i,) , axis=1)
```

```python
df_scores.head()
```

##  QUESTION 3

### Load and read in department Data Fram

```python
department = pd.read_csv("departments.csv", sep=";")
```

### Merging the personality score and the department data frame

```python
#merging department with personality scores
score_dep = pd.merge(department, df_scores, how='left')
score_dep.head()
```

### Assert statement to check that the newly created merged data frame has the same amount of rows as the department data frame, and the expected number of columns 

```python
#Assertation/ checking whether the two df's have the same amount of rows & columns 
try:
    assert len(df_scores) == len(score_dep), "The two data frames do not have the same amount of rows & columns" 
    print('True')
except AssertionError as msg:
    print(msg)
```

## QUESTION 4

### Filter the merged data frame so that you get only the applicants who scored less than 30 on emotional stability, conscientiousness AND agreeableness

```python
#Conditioning the merged dataframe
score = score_dep[(score_dep['Conscientiousness']<30) & (score_dep['Agreeableness']<30) & (score_dep['Emotional Stability'] <30)]
score
```

### Print the ID numbers and departments of these applicants to the screen

```python
#Printing ID numbers and departments of these applicants
sep = score[["ID","Department"]]
print(sep)
```

### assigning these applicants the tag “high_risk” in a new column and all the other applicants get the tag “low_risk

```python
#Assigning High risk and low risk 
score_dep['Risk']=np.where((score_dep['Conscientiousness']<30) & (score_dep['Agreeableness']<30) & (score_dep['Emotional Stability'] <30),'high risk','low risk')

score_dep
```

## Question 5

### Create a new data frame with a count of the number of low and high risk applicants within each department.

```python
# apply groupby, and count and unstack data frame
score_new = score_dep.groupby(["Risk", "Department"])["Department"].count().unstack()

# convert to new data frame
score_new = pd.DataFrame(score_new)
score_new
```
